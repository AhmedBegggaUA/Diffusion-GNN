```python
import torch
import numpy as np
import warnings
import warnings
warnings.filterwarnings("ignore")
from torch_geometric.datasets import Planetoid,WebKB,Actor,WikipediaNetwork, LINKXDataset
from torch_geometric.utils import to_dense_adj, to_networkx
dataset = Planetoid(root='./data',name='Cora')
data = dataset[0]
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

# Crear un grafo de ejemplo (reemplaza esto con tu propio grafo)
#G = nx.erdos_renyi_graph(100, 0.15)
G = to_networkx(dataset[0])  
# Calcular los grados de los nodos
degrees = [deg for _, deg in G.degree()]

# Calcular la distribución de grado
degree_values, degree_distribution = np.unique(degrees, return_counts=True)

# Convertir los valores a escala logarítmica (evitando los valores 0 para evitar errores en el logaritmo)
log_degrees = np.log10(degree_values + 1)
log_distribution = np.log10(degree_distribution + 1)

# Ajustar una recta utilizando ajuste lineal (regresión lineal) en la escala log-log
fit = np.polyfit(log_degrees, log_distribution, 1)
fit_fn = np.poly1d(fit)

# Crear el gráfico
plt.figure(figsize=(10, 6))
plt.scatter(log_degrees, log_distribution, label='Distribución de Grado')
plt.plot(log_degrees, fit_fn(log_degrees), color='red', label=f'Ajuste: {fit[0]:.2f}x + {fit[1]:.2f}')
plt.xlabel('Log(Degrees)')
plt.ylabel('Log(Degree Distribution)')
plt.title('Distribución de Grado en Escala Log-Log')
plt.legend()

# Mostrar el gráfico
plt.show()


```
