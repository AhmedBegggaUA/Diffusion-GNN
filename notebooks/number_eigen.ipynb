```python
import numpy as np
import matplotlib.pyplot as plt

#Women = [115, 215, 250, 200]
#Men = [114, 230, 510, 370]
Texas = [74.03,88.30, 91.90]
Cornell = [76.92, 88.01,91.31]
Wisconsin = [70.64,89.17,91.85]
n=3
r = np.arange(n)
width = 0.15

plt.figure(figsize=(10,5))
# Place the values above the bar in the plot
plt.bar(r, Texas, color = 'tab:blue',
		width = width,
		label='Texas',tick_label=['3','8','30'])
plt.bar(r + width, Cornell, color = 'tab:orange',
		width = width, 
		label='Cornell')
plt.bar(r + 2*width, Wisconsin, color = 'tab:green',
		width = width,
		label='Wisconsin')

plt.xlabel("Number of eigenfunctions",fontsize=14,color='black')
plt.ylabel("AUC",fontsize=14,color='black')
plt.title("AUC by number of eigenfunctions",fontsize=16,color='black')

# plt.grid(linestyle='--')
plt.xticks(r + width,['3','8','30'],fontsize=14,color='black')
plt.yticks(np.arange(65, 100, 5))
plt.yticks(fontsize=14,color='black')
plt.legend()
# Plot from the 70 to 100
plt.ylim(65,100)
plt.savefig('irregular_graphs.pdf',dpi=600)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

#Women = [115, 215, 250, 200]
#Men = [114, 230, 510, 370]
Cora = [87.20,90.42, 98.90]
# reverse the order of the list
Cora.reverse()
Citeseer = [88.40, 91.03,98.27]
Citeseer.reverse()
Pubmed = [89.59,94.22,97.31]
Pubmed.reverse()
n=3
r = np.arange(n)
width = 0.15

plt.figure(figsize=(10,5))
# Place the values above the bar in the plot
plt.bar(r, Cora, color = 'tab:blue',
		width = width,
		label='Cora',tick_label=['3','8','30'])
plt.bar(r + width, Citeseer, color = 'tab:orange',
		width = width, 
		label='Citeseer')
plt.bar(r + 2*width, Pubmed, color = 'tab:green',
		width = width,
		label='Pubmed')

plt.xlabel("Number of eigenfunctions",fontsize=14,color='black')
plt.ylabel("AUC",fontsize=14,color='black')
plt.title("AUC by number of eigenfunctions",fontsize=16,color='black')

# plt.grid(linestyle='--')
plt.xticks(r + width,['3','8','30'],fontsize=14,color='black')
plt.yticks(np.arange(85, 100, 5))
plt.yticks(fontsize=14,color='black')
plt.legend()
# Plot from the 70 to 100
plt.ylim(85,100)
plt.savefig('regular_graphs.pdf',dpi=600)
plt.show()
```
